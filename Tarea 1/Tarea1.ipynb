## Tarea 1
La primer tarea consistirá en programar una **función** que genere señales senoidales y que permita parametrizar:

#### Parámetros:

vmax : float -> Amplitud máxima de la senoidal (V)

dc   : float -> Valor medio (V)

f0   : float -> Frecuencia de la señal (Hz)

ph   : float -> Fase inicial (radianes)

nn   : int   -> Cantidad de muestras

fs   : float -> Frecuencia de muestreo (Hz)

#### Retorna:

tt : numpy array -> Vector de tiempos

yy : numpy array -> Valores de la señal en cada instante de tiempo


### 1: Importo las librerías necesarias 

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 2: Creo mi función

```python
def crear_sin(vmax, dc, f0, ph, nn, fs):
    Ts = 1 / fs  # Periodo de muestreo
    Ttotal = nn * Ts  # Duración total de la señal

    # Crear el vector de tiempos / el false indica que no incluye el ultimo punto
    tt = np.linspace(0, Ttotal, nn, endpoint=False)

    # Generar la señal senoidal
    yy = dc + vmax * np.sin(2 * np.pi * f0 * tt + ph)
    return tt, yy
```

### 3: Hago una prueba

```python
# Prueba general
tt, xx = crear_sin(1, 0, 4, 0.5*np.pi, 50, 100)

# Graficar
plt.figure(figsize=(15, 3))
plt.plot(tt, xx,'m')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (V)")
plt.title("Señal Senoidal Muestreada")
plt.grid()
plt.show()
```

### 4: Realizo los experimentos que se comentaron en clase

#### Gané :)

Respecto a los graficos de $ f_0=999 $ y $f_0=1001$, lo que sucede es que estoy tomando aproximadamente una muestra por ciclo con una especie de retardo o adelanto dependiendo del caso.

En $f_0=999$ el $T_s<T_0$, es decir, voy a muestear un poquito antes de que termine el ciclo. Para entender las consecuencias, grafiqué una señal con la misma $f_0 $ y $T_0$, pero con mayor frecuencia muestral. Además, tomé menos muestras para hacer una especie de zoom en la imagen y ver que sucede.


```python
plt.figure(figsize=(12, 5))
f0 = 999
tt, xx = crear_sin(1, 0, f0, 0, 20, 1000)
plt.plot(tt, xx, 'm-*', label='fs=1000', linewidth=0.7)  # Primera señal con línea más fina

# Agregar señal senoidal con f0 = 50
tt, xx = crear_sin(1, 0, f0, 0, 200, 9990)
plt.plot(tt, xx, 'c-', label='fs=9990',linewidth=0.3)  # Segunda señal

plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (V)")
plt.title("Señales Senoidales f0=999")
plt.legend()  # Mostrar la leyenda para identificar las señales
plt.grid()

# Mostrar la figura completa
plt.tight_layout()
plt.show()
```
