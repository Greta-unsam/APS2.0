```python
import numpy as np
import matplotlib.pyplot as plt


N = 512
w = np.linspace(0, 2 * np.pi, N, endpoint=False)


h_a = np.zeros(N)
h_a[0] = 1
h_a[1] = 1
h_a[2] = 1
h_a[3] = 1

# Sistema b: y(n) = x(n) + x(n-1) + x(n-2) + x(n-3) + x(n-4)
h_b = np.zeros(N)
h_b[0:5] = 1

# Sistema c: y(n) = x(n) - x(n-1)
h_c = np.zeros(N)
h_c[0] = 1
h_c[1] = -1

# Sistema d: y(n) = x(n) - x(n-2)
h_d = np.zeros(N)
h_d[0] = 1
h_d[2] = -1


H_a = np.fft.fft(h_a, N)
H_b = np.fft.fft(h_b, N)
H_c = np.fft.fft(h_c, N)
H_d = np.fft.fft(h_d, N)


sistemas = {
    "Sistema a": H_a,
    "Sistema b": H_b,
    "Sistema c": H_c,
    "Sistema d": H_d
}

for nombre, H in sistemas.items():
    plt.figure(figsize=(12, 4))
    
    # Módulo
    plt.subplot(1, 2, 1)
    plt.plot(w, np.abs(H))
    plt.title(f"{nombre} - Módulo")
    plt.xlabel("Frecuencia (rad/muestra)")
    plt.ylabel("|H(w)|")
    plt.grid(True)
    
    # Fase
    plt.subplot(1, 2, 2)
    plt.plot(w, np.angle(H))
    plt.title(f"{nombre} - Fase")
    plt.xlabel("Frecuencia (rad/muestra)")
    plt.ylabel("∠H(w) [rad]")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
```
